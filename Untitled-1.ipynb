```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from retinaface import RetinaFace
```

```python
#part0
```

```python
img = cv2.imread("D:/Project/Python/trains/Crowley.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

```

```python
blured_image = cv2.GaussianBlur(img, (51,51), -1)
plt.imshow(blured_image)
```

```python
Identity = np.array([[0, 0, 0],
                    [0, 1, 0],
                    [0, 0, 0]])


filter1 = cv2.filter2D(src=img, ddepth=-1, kernel=Identity)

cv2.imshow('identity', filter1)
```

```python
plt.imshow(filter1)
```

```python
Left_Sobel = np.array([[1, 0, -1],
                    [2, 0, -2],
                    [1, 0, -1]])


filter2 = cv2.filter2D(src=img, ddepth=-1, kernel=Left_Sobel)

cv2.imshow('Left Sobel', filter2)
```

```python
plt.imshow(filter2)
```

```python
Random = np.array([[-2, -1, 0],
                    [-1 , 1 , 1],
                    [0, 1 , 2]])


filter3 = cv2.filter2D(src=img, ddepth=-1, kernel=Random)

cv2.imshow('Random', filter3)
```

```python
plt.imshow(filter3)
```

```python
pth = "D:/Project/Python/trains/git image/faces/"
all_img = os.listdir(pth)
print(all_img)
print(len(all_img))
```

```python
index = np.random.randint(0 , len(all_img))
img = cv2.imread(f'{pth}{all_img[index]}')
img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(img)

```

```python
index = np.random.randint(0 , len(all_img))
img = cv2.imread(f'{pth}{all_img[index]}')

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

lower_bound = np.array([0,0,0])
upper_bound = np.array([255,255,255])

mask = cv2.inRange(hsv , lower_bound, upper_bound)

gray_pix = cv2.countNonZero(mask)

grayy = gray_pix/(mask.shape[0] * mask.shape[1])


print(grayy)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
